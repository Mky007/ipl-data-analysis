```python
mtch['season']=pd.DatetimeIndex(mtch['date']).year
mtch.head()
```

```python
mtch_season=mtch.groupby(['season'])['id'].count().reset_index().rename(columns={'id':'matches'})
mtch_season
```

```python
sn.countplot(mtch['season'])
#sn.lineplot(data=mtch_season)
plt.xticks(rotation=45)
plt.title("Total matches played",fontweight='bold')
```

```python
season_data=mtch[['id','season']].merge(ball, left_on = 'id', right_on = 'id', how = 'left').drop('id', axis = 1)
season_data.head()
```

```python
season=season_data.groupby(['season'])['total_runs'].sum().reset_index()
p=season.set_index('season')
ax = plt.axes()
ax.set(facecolor = "black")
sn.lineplot(data=p,palette="magma") 
plt.title('Total runs in each season',fontsize=12,fontweight="bold")
#plt.show
```

```python
runs_per_season=pd.concat([mtch_season,season.iloc[:,1]],axis=1)
runs_per_season['Runs scored per match']=runs_per_season['total_runs']/runs_per_season['matches']
runs_per_season.set_index('season',inplace=True)
runs_per_season
```

```python
toss=mtch['toss_winner'].value_counts()
ax = plt.axes()
ax.set(facecolor = "grey")
sn.set(rc={'figure.figsize':(10,5)},style='darkgrid')
ax.set_title('No. of tosses won by each team',fontsize=15,fontweight="bold")
sn.barplot(y=toss.index, x=toss, orient='h',palette="icefire",saturation=1)
plt.xlabel('# of tosses won')
plt.ylabel('Teams')
plt.show()
```
